### Styles Graphiques
Il existe beaucoup de styles a ajouter aux graphiques. Voici les plus importants a retenir :
- **c** : couleur de la ligne
- **lw** : epaisseur de la ligne (pour les graphiques plot)
- **ls** : style de la ligne (pour les graphiques plot)
- **size** : taille du point (pour les graphiques scatter **seul**)
- **marker** : style de points (pour les graphiques scatter)
- **alpha** : transparence du graphique


```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(0, 2, 10)
y = X**2

plt.plot(X, y, c='green', lw=1, ls='--', marker='o', alpha=0.8)  #D : diamond
plt.show()
```

## 2. Cycle de vie d'une figure
Pour créer des figures proprements, on doit suivre le cycle de vie suivant :
1. **plt.figure**(figsize=(a,b))
2. **plt.plot()**
3. optional **Extras (titre, axes, legendes)**
4. optional **plt.save()**
5. **plt.show()**

```python
X = np.linspace(0, 2, 10)

plt.figure(figsize=(8, 6)) # Création d'une figure
plt.plot(X, X**2, label='quadratique') # premiere courbe
plt.plot(X, X**3, label='cubique') # deuxieme courbe
# Autres informations
plt.title('figure 1') # titre
plt.xlabel('axe x') # axes
plt.ylabel('axe y') # axes
plt.legend() # legend, takes what's in label

# plt.savefig('figure.png') # sauvegarde la figure dans le repertoire de travail
plt.show() # affiche la figure
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 8, 7, 11]

plt.figure(figsize=(5, 4))

plt.scatter(x, y, label='Données', color='blue', marker='o')  #marker= o, *, ^, etc.

plt.xlabel('axe x')
plt.ylabel('axe y')
plt.title('Titre')
plt.legend()
plt.show()
```

## 4.2. Box plot

```python
data = [1, 1.4, 2.3, 3, 3.8, 4.2, 5, 6.3, 8, 12]
len(data)
```

```python
import numpy as np

Q1 = np.percentile(data, 25)          #by default, method=linear    (check via shift+tab)
Q3 = np.percentile(data, 75)

print(Q1, '\t', Q3)
```

```python

IQR = Q3 - Q1

box_length = IQR
print("La taille de l'IQR est :", box_length)
```

```python
print(data[4], data[5])  #----> mediane = 4  data = [1,   1.4,   2.3,   3,   3.8,   4,   4.2,   5,   6.3,   8,   12]
```

```python
plt.figure(figsize=(5, 4))

plt.boxplot(data, vert=False)     #vert : vertical    #shift+tab pour voir des détails importants

plt.title('Box Plot Example')
plt.show()
```

## 4.3 Histogram

**distribution discontinue des données**

* Pour chaque valeur ou catégories --> combien elle a été répétée

```python
## 30 se répéte trois fois, 55 se répéte deux fois

data = [23, 30, 30, 30, 37, 40, 45, 50, 55, 55, 60, 65, 70, 75, 80, 85]

len(data)
```

```python
plt.figure(figsize=(5, 4))

plt.hist(data, bins=len(data),   ## bins : combien de catégories pour diviser la distribution
         edgecolor='black', 
         alpha=0.7, color='skyblue')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.randint(0, 255, [50, 50])  
print(np.max(X), np.min(X))
```

```python
# Afficher l'image
plt.figure(figsize=(12, 3))
plt.subplot(1, 2, 1)
plt.imshow(X, cmap='gray')
plt.title('Image en niveaux de gris aléatoire')

# Afficher l'histogramme de l'image
plt.subplot(1, 2, 2)
plt.hist(X.ravel(), bins=50, color='gray', edgecolor='black')
plt.title('Histogramme de l\'image')

plt.tight_layout()
plt.show()
```

La bibliothèque Matplotlib est très riche et organisée en plusieurs sous-modules pour permettre la création de graphiques 2D et 3D en Python. Voici une vue d'ensemble de la structure générale de Matplotlib :

### 1. **`matplotlib.pyplot`** : Interface de haut niveau
Le sous-module `pyplot` est celui le plus utilisé car il fournit une interface simple pour créer et manipuler des graphiques. Il contient des fonctions de haut niveau similaires à celles du fameux langage MATLAB.

- **Fonctionnalités principales** :
  - **`plot()`** : Pour tracer des graphiques 2D.
  - **`scatter()`** : Pour tracer des diagrammes de dispersion.
  - **`bar()`** : Pour créer des diagrammes en barres.
  - **`hist()`** : Pour générer des histogrammes.
  - **`show()`** : Pour afficher le graphique.
  - **`figure()`** : Pour créer une nouvelle figure (fenêtre graphique).

### 2. **`matplotlib.figure`** : Gestion des figures
Le module `figure` gère la création et la manipulation de figures (les fenêtres ou surfaces sur lesquelles les graphiques sont tracés).

- **Classes principales** :
  - **`Figure`** : Représente la fenêtre graphique ou la "figure" dans laquelle tu traces des graphes.
  - **`Subplot`** : Permet de diviser une figure en plusieurs sous-graphes.

### 3. **`matplotlib.axes` et `matplotlib.axis`** : Gestion des axes et des échelles
Les axes représentent le système de coordonnées dans lequel les données sont tracées, tandis que les échelles (axes) déterminent les plages de valeurs.

- **Classes principales** :
  - **`Axes`** : Un des objets les plus importants. Il s'agit d'un graphique dans la figure.
  - **`Axis`** : Contrôle les limites et l'étiquetage des axes (x et y).
  - **`Subplot`** : Un `Axes` spécifique qui est créé dans une figure divisée en plusieurs sous-graphes.

### 4. **`matplotlib.lines`** : Tracer des lignes
Ce module gère la création et le style des lignes dans les graphiques.

- **Classes principales** :
  - **`Line2D`** : Représente une ligne dans un graphique, souvent utilisée implicitement par des fonctions comme `plot()`.

### 5. **`matplotlib.patches`** : Formes géométriques
Ce module contient des classes pour dessiner des formes géométriques comme des cercles, des rectangles, etc.

- **Classes principales** :
  - **`Patch`** : Classe de base pour les formes.
  - **`Rectangle`, `Circle`, `Polygon`** : Formes spécifiques.

### 6. **`matplotlib.text`** : Gestion du texte
Ce module permet d'ajouter du texte aux graphiques, par exemple des titres, des labels, ou des annotations.

- **Classes principales** :
  - **`Text`** : Représente du texte dans un graphique.
  - **`Annotation`** : Utilisé pour annoter un point spécifique dans un graphique.

### 7. **`matplotlib.image`** : Affichage d'images
Ce module gère le chargement et l'affichage d'images dans les graphiques.

- **Classes principales** :
  - **`AxesImage`** : Pour afficher une image dans les axes.
  - **`imshow()`** : Fonction pour afficher des images.

### 8. **`mpl_toolkits.mplot3d`** : Visualisation 3D
Pour les graphiques 3D, Matplotlib propose une extension via le module `mplot3d`. Ce module permet de tracer des surfaces 3D, des diagrammes de dispersion en 3D, etc.

- **Classes principales** :
  - **`Axes3D`** : Pour créer et manipuler des graphiques 3D.
  
### 9. **`matplotlib.animation`** : Animation
Matplotlib permet aussi de créer des animations en manipulant les objets d'une figure de manière dynamique.

- **Classes principales** :
  - **`FuncAnimation`** : Pour animer des graphiques.
  - **`ArtistAnimation`** : Permet d'animer des objets d'art dans une figure.

### 10. **`matplotlib.colors` et `matplotlib.cm`** : Gestion des couleurs
Ces modules gèrent la manière dont les couleurs sont utilisées et affichées dans les graphiques, ainsi que les cartes de couleurs (color maps).

- **Fonctionnalités principales** :
  - **`ListedColormap`, `LinearSegmentedColormap`** : Pour créer et manipuler des cartes de couleurs.
  - **`colormaps()`** : Liste des cartes de couleurs prédéfinies.
  
### 11. **`matplotlib.backends`** : Backends pour le rendu
Matplotlib supporte plusieurs *backends*, qui sont des moteurs de rendu qui permettent d'afficher ou de sauvegarder les graphiques dans différents formats (fichiers, écrans interactifs, etc.).

- **Exemples de backends** :
  - **`TkAgg`** : Pour une interface graphique avec Tkinter.
  - **`Agg`** : Backend pour le rendu en mode image raster (PNG).
  - **`PDF`** : Pour exporter des graphiques au format PDF.

---

### Schéma de fonctionnement :

1. **Figure** (fenêtre) : Contient une ou plusieurs **Axes** (graphiques).
2. **Axes** : Représentent un espace de coordonnées (x, y ou x, y, z) où sont tracées des données.
3. **Artist** : Tout élément graphique dans une figure est un `Artist` (lignes, texte, images, etc.).
